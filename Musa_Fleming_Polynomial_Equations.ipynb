```python
import numpy as np
import matplotlib.pyplot as plt

def poly():
  try:
    equation = input("Enter polynomial equation using x as the variable: ")

    x = np.linspace(-10, 10, 400)
    y = eval(equation)

    plt.plot(x, y)
    plt.title(f"Graph of {equation}")
    plt.xlabel("X Axis")
    plt.ylabel("Y Axis")
    plt.grid(True)
    plt.show

  except:
    print("Please enter a valid polynomial equation")

poly()
```
